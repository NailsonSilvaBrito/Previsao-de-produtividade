```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Correlação temperatura média x produtividade
correlacao_temp = dados_limpos['Temp_média_(°c)'].corr(dados_limpos['Produtividade_(t/ha)'])
print(f"Correlação temperatura média x produtividade: {correlacao_temp:.2f}")

# Gráfico com ajuste polinomial
x = dados_limpos['Temp_média_(°c)']
y = dados_limpos['Produtividade_(t/ha)']
poly = np.poly1d(np.polyfit(x, y, 2))

plt.figure(figsize=(8,6))
sns.scatterplot(x=x, y=y, alpha=0.7)
plt.plot(np.sort(x), poly(np.sort(x)), color='red')
plt.title("Produtividade vs Temperatura média")
plt.show()
```
